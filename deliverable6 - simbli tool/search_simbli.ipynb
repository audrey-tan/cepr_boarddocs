Code must be ran in Chrome console after using the search feature of Simbli. (https://simbli.eboardsolutions.com/Index.aspx?S=2)

```python
# For ONE school

(async () => {
  const buttons = document.querySelectorAll('button.btnTrans.text-left');
  if (buttons.length < 1) {
    console.error("Less than one document button found.");
    return;
  }

  const firstButton = buttons[0];
  console.log("Clicking the first button:", firstButton.textContent.trim());
  firstButton.click();

  await new Promise(resolve => setTimeout(resolve, 1500));

  while (true) { 
    const modal = document.querySelector('.modal, .mat-dialog-container, [role="dialog"], .modal-content');
    if (!modal) {
      console.warn("Modal not found.");
      return;
    }

    console.log(`Modal opened. Searching for yellow-highlighted PDF (iteration ${i + 1})`);

    // Try finding a yellow-highlighted PDF via class
    let yellowLink = modal.querySelector('a.bgYellow[href*=".aspx"]');

    // If not found, try finding via inline style background color
    if (!yellowLink) {
      const yellowStyled = modal.querySelector('[style*="background-color: rgb(255, 255, 0)"]');
      if (yellowStyled) {
        yellowLink = yellowStyled.querySelector('a[href*=".aspx"]');
      }
    }

    if (yellowLink) {
      const pdfUrl = new URL(yellowLink.getAttribute('href'), window.location.origin).toString();
      const fileName = (yellowLink.getAttribute('title') || yellowLink.textContent.trim())
        .replace(/[\/\\:*?"<>|]/g, '_') + '.pdf';

      const a = document.createElement('a');
      a.href = pdfUrl;
      a.download = fileName;
      a.target = '_blank';
      document.body.appendChild(a);
      a.click();
      document.body.removeChild(a);

      console.log(`✅ Download triggered for yellow-highlighted file: "${fileName}"`);
    } else {
      console.log("No yellow-highlighted item found in this modal.");
    }

    // Move to the next item using the right-chevron button
    const chevronButton = modal.querySelector('i.fa-chevron-right');
    if (chevronButton) {
      const chevronParentButton = chevronButton.closest('button');
      if (chevronParentButton) {
        chevronParentButton.click();
        console.log("Moved to next item.");
        await new Promise(resolve => setTimeout(resolve, 1500));
      } else {
        console.warn("Chevron icon found but parent <button> is missing.");
        break;
      }
    } else {
      console.warn("No more right chevrons. Ending loop.");
      break;
    }
  }
})();
```

```python
# For ALL schools

(async () => {
  // Step 1: Expand the first state
  const stateButtons = document.querySelectorAll('td.col-state button.btnTrans');
  if (stateButtons.length === 0) {
    console.warn("No state buttons found.");
    return;
  }
  console.log("Expanding first state...");
  stateButtons[0].click();
  await new Promise(resolve => setTimeout(resolve, 1000));

  // Step 2: Expand the first school
  const schoolButtons = document.querySelectorAll('td.col-org button.btnTrans');
  if (schoolButtons.length === 0) {
    console.warn("No school buttons found under expanded state.");
    return;
  }
  console.log("Expanding first school...");
  schoolButtons[0].click();
  await new Promise(resolve => setTimeout(resolve, 1000));

  // Step 3: Click the first meeting item button like "School Handbooks"
  const itemButtons = document.querySelectorAll('td.col-it button.btnTrans.text-left');
  if (itemButtons.length === 0) {
    console.error("No item buttons found under expanded school.");
    return;
  }
  console.log("Clicking first item button:", itemButtons[0].textContent.trim());
  itemButtons[0].click();
  await new Promise(resolve => setTimeout(resolve, 1500));

  // Step 4: Handle modal and download all yellow-highlighted PDFs
  while (true) {
    const modal = document.querySelector('.modal, .mat-dialog-container, [role="dialog"], .modal-content');
    if (!modal) {
      console.warn("Modal not found.");
      return;
    }

    console.log(`Modal found (iteration ${i + 1}). Searching for yellow-highlighted PDFs...`);

    // Find all yellow-highlighted links inside the modal
    let yellowLinks = modal.querySelectorAll('a.bgYellow[href*=".aspx"]');

    // If not found, try finding via inline style background color
    if (yellowLinks.length === 0) {
      const yellowStyled = modal.querySelectorAll('[style*="background-color: rgb(255, 255, 0)"]');
      yellowLinks = [];
      yellowStyled.forEach(styledElement => {
        const link = styledElement.querySelector('a[href*=".aspx"]');
        if (link) yellowLinks.push(link);
      });
    }

    if (yellowLinks.length > 0) {
      yellowLinks.forEach(yellowLink => {
        const pdfUrl = new URL(yellowLink.getAttribute('href'), window.location.origin).toString();
        const fileName = (yellowLink.getAttribute('title') || yellowLink.textContent.trim())
          .replace(/[\/\\:*?"<>|]/g, '_') + '.pdf';

        const a = document.createElement('a');
        a.href = pdfUrl;
        a.download = fileName;
        a.target = '_blank';
        document.body.appendChild(a);
        a.click();
        document.body.removeChild(a);

        console.log(`✅ Download triggered for yellow-highlighted file: "${fileName}"`);
      });
    } else {
      console.log("No yellow-highlighted item found.");
    }

    // Check if we can move to the next item using the right-chevron button
    const chevronButton = modal.querySelector('i.fa-chevron-right');
    const parentBtn = chevronButton?.closest('button');
    if (parentBtn) {
      parentBtn.click();
      console.log("Moved to next item.");
      await new Promise(resolve => setTimeout(resolve, 1500));
    } else {
      console.warn("No more right chevrons. Ending.");
      break;
    }
  }
})();
```
