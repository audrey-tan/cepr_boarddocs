This notebook creates the Google search queries for each Local Education Agency (LEA) to get its BoardDocs website.

Input:
- `../nces2324.csv`

Output:
- `lea_boarddocs_queries.csv`

```python
import pandas as pd

df = pd.read_csv("../nces2324.csv")
```

```python
# the query should be its name
df['query'] = df['LEA_NAME']
```

```python
df['query'].duplicated(keep=False).sum()
```

```python
# for duplicated ones, we add in the zip code
# we don't do it for all queries because some school don't put their addresses on the boarddocs website

df.loc[df['query'].duplicated(keep=False), 'query'] = df['LEA_NAME'] + ' ' + df['LZIP'].astype(str)
```

```python
sum(df['query'].duplicated(keep=False))
```

```python
df.loc[df['query'].duplicated(keep=False)].sort_values(by="query")
```

```python
# turns out that all 17 duplicated rows have the same zip code too.
# just give the same query, it's fine
# don't include zip code
df['query'] = df['LEA_NAME']
```

```python
# select cols
df = df[["LEA_NAME", "LEAID", "STATENAME"]]
df.to_csv("lea_boarddocs_queries.csv", index=False)
```
