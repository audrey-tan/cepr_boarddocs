```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://meetings.boardbook.org/Public/"

response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    school_html_lines = soup.find_all('a', href=True)

    schools = []

    for line in school_html_lines:
        href = line.get('href')

        if '/Public/Organization/' in href:
            index = href.split('/')[-1] # Get school index
            school_name = line.get_text(strip=True) # Get school name
            schools.append([school_name, index])

    with open('premier_boardbook_schools.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['School Name', 'Index'])
        writer.writerows(schools)
```

```python
# Compare results from BoardBook Manual and BoardBook Premier to see if it is the same

import csv

def compare_schools(file1, file2):
    # Read both CSV files into dictionaries with school name as key and index as value
    schools1 = {}
    schools2 = {}

    with open(file1, mode='r', newline='', encoding='utf-8') as f1, \
         open(file2, mode='r', newline='', encoding='utf-8') as f2:
        
        reader1 = csv.reader(f1)
        reader2 = csv.reader(f2)

        # Skip the header row if present
        next(reader1)  # Skip header in file 1
        next(reader2)  # Skip header in file 2

        # Read the school names and indexes from file 1
        for row in reader1:
            school_name = row[0]  # Assuming the school name is in the first column
            index = row[1]  # Assuming the index is in the second column
            schools1[school_name] = index

        # Read the school names and indexes from file 2
        for row in reader2:
            school_name = row[0]  # Assuming the school name is in the first column
            index = row[1]  # Assuming the index is in the second column
            schools2[school_name] = index

    # Compare the schools in both files
    all_schools = set(schools1.keys()).union(set(schools2.keys()))  # All unique school names from both files

    for school in all_schools:
        index1 = schools1.get(school)
        index2 = schools2.get(school)

        if index1 is None:
            print(f"{school} is missing in {file1}")
        elif index2 is None:
            print(f"{school} is missing in {file2}")
        elif index1 != index2:
            print(f"Index mismatch for {school}: {file1} has {index1}, {file2} has {index2}")


compare_schools('boardbook_schools.csv', 'premier_boardbook_schools.csv')
```

```python
import csv

# Input CSV file (replace with your actual file path)
input_file = "filtered_schools.csv"

# Output TS file
output_file = "boardbookData.ts"

data = []

with open(input_file, newline="", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        school_name = row["School Name"].replace('"', '\\"')
        index = row["Index"]
        data.append(f'  {{ school_name: "{school_name}", index: "{index}", leaid: "0" }}')

with open(output_file, "w", encoding="utf-8") as f:
    f.write("export const boardbookData = [\n")
    f.write(",\n".join(data))
    f.write("\n];\n")

print(f"Saved to {output_file}")
```
